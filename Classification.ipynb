```python
def change_value(X,Y):
    number = []
    for key,values in data["projectstatusdisplay"].iteritems():
        if(values) == X:
            number.append(key)
            
    for i in number:
        if data["projectstatusdisplay"][i] == X:
            data["projectstatusdisplay"][i] = Y
```

```python
values = ['Dropped','Pipeline']
for i in values:
    change_value(i,'Closed')
```
